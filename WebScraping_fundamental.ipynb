**Import all the above packages.**

```
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

**Send request to amazon website**

```
url = "https://www.amazon.it/s?k=books&__mk_it_IT=%C3%85M%C3%85%C5%BD%C3%95%C3%91&crid=1UPARTD6PUDCH&sprefix=books%2Caps%2C90&ref=nb_sb_noss_1"
```

```
# Header for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36','Accept-Language':'en-Us,en;q=0.5'})
```

```
webpage = requests.get(url , headers = HEADERS)
```

```
# At this step the request is returning the document in bytes format.
type(webpage.content)
```

```
# To change it to html document we need to parse it using beautifulSoup
soup = BeautifulSoup(webpage.content , 'html.parser')
```

```
# We get a BeautifulSoup object
type(soup)
```

```
# Fetch links as list of tag objects
links =  soup.find_all("a" , attrs = {'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'}) 
```

```
links
```

```
# Fetch the href link and then append the amazon website link to it to make a proper link
link =links[0].get("href")
product_list = "https://amazon.it/" + link
```

```
product_list
```

**Now we need to again make the request to the new website page that we have stored in the product_list variable**

```
new_webpage = requests.get(product_list , headers = HEADERS)
```

```
new_soup = BeautifulSoup(new_webpage.content , 'html.parser')
type(new_soup)
```

```
new_soup.prettify()
```

```
new_soup.find("span", attrs= { 'id':'productTitle'}).text.strip()
```
